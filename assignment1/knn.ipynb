kNN分类器包含两个阶段:

- 训练阶段，分类器获取训练数据并简单地记住它。
- 测试阶段, kNN将测试图像与所有训练图像进行比较，并计算出前k个最相似的训练示例的标签来对每个测试图像进行分类。

**问题 2**

我们还可以使用其他距离指标，例如L1距离。

记图像 $I_k$ 的每个位置 $(i,j)$ 的像素值为 $p_{ij}^{(k)}$，

所有图像上的所有像素的均值 $\mu$ 为 

$$\mu=\frac{1}{nhw}\sum_{k=1}^n\sum_{i=1}^{h}\sum_{j=1}^{w}p_{ij}^{(k)}$$

并且所有图像的每个像素的均值 $\mu_{ij}$ 为

$$\mu_{ij}=\frac{1}{n}\sum_{k=1}^np_{ij}^{(k)}.$$

标准差 $\sigma$ 以及每个像素的标准差 $\sigma_{ij}$ 的定义与之类似。

以下哪个预处理步骤不会改变使用L1距离的最近邻分类器的效果？选择所有符合条件的答案。
1. 减去均值 $\mu$ ($\tilde{p}_{ij}^{(k)}=p_{ij}^{(k)}-\mu$.)
2. 减去每个像素均值 $\mu_{ij}$  ($\tilde{p}_{ij}^{(k)}=p_{ij}^{(k)}-\mu_{ij}$.)
3. 减去均值 $\mu$ 然后除以标准偏差 $\sigma$.
4. 减去每个像素均值 $\mu_{ij}$ 并除以每个素标准差 $\sigma_{ij}$.
5. 旋转数据的坐标轴。

$\color{blue}{\textit 你的回答:}$
    步骤1、2、3不会改变使用L1距离的最近邻分类器的效果，步骤4会改变使用L1距离的最近邻分类器的效果。若考虑理想情况，步骤5不会改变使用L1距离的最近邻分类器的效果，而若考虑实际情况，步骤5不会改变使用L1距离的最近邻分类器的效果当且仅当旋转的角度为 $\frac{\pi}{2}$ 的整数倍时。

$\color{blue}{\textit 你的解释:}$


当使用L1距离作为k-NN的距离指标时，两张图片的距离公式为：

$$dist(k_1, k_2) = \sum_{i=1}^h \sum_{j=1}^w |p_{ij}^{(k_1)} - p_{ij}^{(k_2)}|$$

#### 1. 减去均值 $\mu$ ($\tilde{p}_{ij}^{(k)}=p_{ij}^{(k)}-\mu$)

$$
\begin{align}
dist'(k_1, k_2) &= \sum_{i=1}^h \sum_{j=1}^w |(p_{ij}^{(k_1)}-\mu) - (p_{ij}^{(k_2)}-\mu)|\\
           &= \sum_{i=1}^h \sum_{j=1}^w |p_{ij}^{(k_1)} - p_{ij}^{(k_2)}|\\
           &= dist(k_1, k_2)
\end{align}
$$

因此该预处理步骤不会改变使用L1距离的最近邻分类器的效果。
   
#### 2. 减去每个像素均值 $\mu_{ij}$  ($\tilde{p}_{ij}^{(k)}=p_{ij}^{(k)}-\mu_{ij}$)

$$
\begin{align}
dist'(k_1, k_2) &= \sum_{i=1}^h \sum_{j=1}^w |(p_{ij}^{(k_1)}-\mu_{ij}) - (p_{ij}^{(k_2)}-\mu_{ij})|\\
           &= \sum_{i=1}^h \sum_{j=1}^w |p_{ij}^{(k_1)} - p_{ij}^{(k_2)}|\\
           &= dist(k_1, k_2)
\end{align}
$$

因此该预处理步骤不会改变使用L1距离的最近邻分类器的效果。

#### 3. 减去均值 $\mu$ 然后除以标准偏差 $\sigma$.

$$
\begin{align}
dist'(k_1, k_2) &= \sum_{i=1}^h \sum_{j=1}^w |\frac{p_{ij}^{(k_1)}-\mu}{\sigma} - \frac{p_{ij}^{(k_2)}-\mu}{\sigma}|\\
           &= \frac{1}{\sigma}\sum_{i=1}^h \sum_{j=1}^w |p_{ij}^{(k_1)} - p_{ij}^{(k_2)}|\\
           &= \frac{1}{\sigma}dist(k_1, k_2)
\end{align}
$$

相当于所有图片的距离计算都多乘了一个相同的正数 $\frac{1}{\sigma}$，这对距离比较并没有影响，大的还是大，小的还是小。因此该预处理步骤不会改变使用L1距离的最近邻分类器的效果。

#### 4. 减去每个像素均值 $\mu_{ij}$ 并除以每个像素标准差 $\sigma_{ij}$.

$$
\begin{align}
dist'(k_1, k_2) &= \sum_{i=1}^h \sum_{j=1}^w |\frac{p_{ij}^{(k_1)}-\mu_{ij}}{\sigma_{ij}} - \frac{p_{ij}^{(k_2)}-\mu_{ij}}{\sigma_{ij}}|\\
           &= \sum_{i=1}^h \sum_{j=1}^w \frac{1}{\sigma_{ij}} |p_{ij}^{(k_1)} - p_{ij}^{(k_2)}|
\end{align}
$$

相当于两张图片的L1距离变成了每个像素点L1距离的加权求和，权重为 $\frac{1}{\sigma_{ij}}$，与像素点的位置有关。这样会导致距离的比较出现不一样的结果。假设图片 $k_1$ 和 $k_2$ 在第 $i$ 行第 $j$ 列像素值完全一样，而图片 $k_1$ 和 $k_3$ 在该位置的像素值差值为 $d$，那么在原来的距离计算公式中，该像素点对图片 $k_1$ 和 $k_3$ 的总距离的贡献为 $d$；而采用了上述数据预处理后，该像素点对图片 $k_1$ 和 $k_3$ 的总距离的贡献为 $\frac{d}{\sigma_{ij}}$ ，而对图片 $k_1$ 和 $k_2$ 的总距离贡献依旧为0。因此，该预处理会影响每个像素点的距离在总距离中的占比，这将导致图片间距离关系发生变化，从而影响k-NN算法的结果。

#### 5. 旋转数据的坐标轴。

$$
\begin{align}
dist'(k_1, k_2) &= \sum_{(i',j')} |p_{i'j'}^{(k_1)} - p_{i'j'}^{(k_2)}|\\
           &\overset{ideal}{=} \sum_{(i,j)} |p_{ij}^{(k_1)} - p_{ij}^{(k_2)}|\\
           &= dist(k_1, k_2)
\end{align}
$$

其中 $(i',j')$ 是旋转坐标轴后的坐标， $(i,j)$ 是 $(i',j')$ 对应的旋转前的坐标。在理想情况下， $(i,j)$ 和 $(i',j')$ 是一一对应的，因此该预处理操作不会影响k-NN算法的结果。但是实际中，$(i,j)$ 和 $(i',j')$极有可能不是一一对应的，因为坐标旋转涉及三角函数的计算，都是浮点数计算，而坐标是整数，因此，旋转意味着要对旋转后的浮点数理论坐标取整处理（通常实践过程与这个相反，也就是说 $p_{i'j'}=p_{ij}, s.t. i=int(cos\theta i'+sin\theta j'),j=int(cos\theta i'-sin\theta j')$，即对旋转前的浮点数理论坐标取整处理，获取对应的像素值），因此必然存在图片数据的损坏，可能有多个旋转后的像素点的像素值对应原图片同一个点的像素值，而原图片中某些像素点的像素值就没有出现在旋转后的图片中了。这样以来，计算图片间的距离就会出现不同的结果，因此会对k-NN算法的结果有影响。图片像素的总数量越大，该影响越小。只有当角度是 $\frac{\pi}{2}$ 的整数倍时，该影响为0，真实结果和理想结果完全一致。


```python
# 现在，通过部分矢量化并且使用单层循环的来加快距离矩阵的计算。
# 需要实现函数compute_distances_one_loop并运行以下代码：

dists_one = classifier.compute_distances_one_loop(X_test)

# 为了确保我们的矢量化实现正确，我们要保证它的结果与最原始的实现方式结果一致。
# 有很多方法可以确定两个矩阵是否相似。最简单的方法之一就是Frobenius范数。 
# 如果您以前从未了解过Frobenius范数，它其实是两个矩阵的所有元素之差的平方和的平方根；
# 换句话说，就是将矩阵重整为向量并计算它们之间的欧几里得距离。

difference = np.linalg.norm(dists - dists_one, ord='fro')
print('One loop difference was: %f' % (difference, ))
if difference < 0.001:
    print('Good! The distance matrices are the same')
else:
    print('Uh-oh! The distance matrices are different')
```

```python
# 现在完成compute_distances_no_loops实现完全矢量化的版本并运行代码
dists_two = classifier.compute_distances_no_loops(X_test)

# 检查距离矩阵是否与我们之前计算出的矩阵一致：
difference = np.linalg.norm(dists - dists_two, ord='fro')
print('No loop difference was: %f' % (difference, ))
if difference < 0.001:
    print('Good! The distance matrices are the same')
else:
    print('Uh-oh! The distance matrices are different')
```

```python
# 让我们比较一下三种实现方式的速度
def time_function(f, *args):
    """
    Call a function f with args and return the time (in seconds) that it took to execute.
    """
    import time
    tic = time.time()
    f(*args)
    toc = time.time()
    return toc - tic

two_loop_time = time_function(classifier.compute_distances_two_loops, X_test)
print('Two loop version took %f seconds' % two_loop_time)

one_loop_time = time_function(classifier.compute_distances_one_loop, X_test)
print('One loop version took %f seconds' % one_loop_time)

no_loop_time = time_function(classifier.compute_distances_no_loops, X_test)
print('No loop version took %f seconds' % no_loop_time)

# 你应该会看到使用完全矢量化的实现会有明显更佳的性能！

# 注意：在部分计算机上，当您从两层循环转到单层循环时，
# 您可能看不到速度的提升，甚至可能会看到速度变慢。
```

### 交叉验证

我们已经实现了kNN分类器，并且可以设置k = 5。现在，将通过交叉验证来确定此超参数的最佳值。

```python
num_folds = 5
k_choices = [1, 3, 5, 8, 10, 12, 15, 20, 50, 100]

X_train_folds = []
y_train_folds = []
################################################################################
# 需要完成的事情: 
# 将训练数据分成多个部分。拆分后，X_train_folds和y_train_folds均应为长度为num_folds的列表，
# 其中y_train_folds [i]是X_train_folds [i]中各点的标签向量。
# 提示：查阅numpy的array_split函数。                             
################################################################################
# *****START OF YOUR CODE (DO NOT DELETE/MODIFY THIS LINE)*****

X_train_folds = np.array_split(X_train, num_folds)
y_train_folds = np.array_split(y_train, num_folds)

# *****END OF YOUR CODE (DO NOT DELETE/MODIFY THIS LINE)*****

# A dictionary holding the accuracies for different values of k that we find when running cross-validation.
# 一个字典，存储我们进行交叉验证时不同k的值的精度。
# 运行交叉验证后，k_to_accuracies[k]应该是长度为num_folds的列表，存储了k值下的精度值。
k_to_accuracies = {}


################################################################################
# 需要完成的事情:  
# 执行k的交叉验证，以找到k的最佳值。
# 对于每个可能的k值，运行k-最近邻算法 num_folds 次，
# 在每次循环下，你都会用所有拆分的数据（除了其中一个需要作为验证集）作为训练数据。
# 然后存储所有的精度结果到k_to_accuracies[k]中。                              
################################################################################
# *****START OF YOUR CODE (DO NOT DELETE/MODIFY THIS LINE)*****

# 交叉验证。有时候，训练集数量较小（因此验证集的数量更小），人们会使用一种被称为
# 交叉验证的方法，这种方法更加复杂些。还是用刚才的例子，如果是交叉验证集，我们就
# 不是取1000个图像，而是将训练集平均分成5份，其中4份用来训练，1份用来验证。然后
# 我们循环着取其中4份来训练，其中1份来验证，最后取所有5次验证结果的平均值作为算
# 法验证结果。

for k in k_choices:
    k_to_accuracies[k] = []
    for i in range(num_folds):
        # prepare training data for the current fold
        X_train_fold = np.concatenate([ fold for j, fold in enumerate(X_train_folds) if i != j ])
        y_train_fold = np.concatenate([ fold for j, fold in enumerate(y_train_folds) if i != j ])
        
        # use of k-nearest-neighbor algorithm
        classifier.train(X_train_fold, y_train_fold)
        y_pred_fold = classifier.predict(X_train_folds[i], k=k, num_loops=0)

        # Compute the fraction of correctly predicted examples
        num_correct = np.sum(y_pred_fold == y_train_folds[i])
        accuracy = float(num_correct) / X_train_folds[i].shape[0]
        k_to_accuracies[k].append(accuracy)

# *****END OF YOUR CODE (DO NOT DELETE/MODIFY THIS LINE)*****

# 打印出计算的精度
for k in sorted(k_to_accuracies):
    for accuracy in k_to_accuracies[k]:
        print('k = %d, accuracy = %f' % (k, accuracy))
```

```python
# 绘制原始观察结果
for k in k_choices:
    accuracies = k_to_accuracies[k]
    plt.scatter([k] * len(accuracies), accuracies)

# 用与标准偏差相对应的误差线绘制趋势线
accuracies_mean = np.array([np.mean(v) for k,v in sorted(k_to_accuracies.items())])
accuracies_std = np.array([np.std(v) for k,v in sorted(k_to_accuracies.items())])
plt.errorbar(k_choices, accuracies_mean, yerr=accuracies_std)
plt.title('Cross-validation on k')
plt.xlabel('k')
plt.ylabel('Cross-validation accuracy')
plt.show()
```

```python
# 根据上述交叉验证结果，为k选择最佳值，使用所有训练数据重新训练分类器，
# 并在测试中对其进行测试数据。您应该能够在测试数据上获得28％以上的准确性。

best_k = k_choices[accuracies_mean.argmax()]

classifier = KNearestNeighbor()
classifier.train(X_train, y_train)
y_test_pred = classifier.predict(X_test, k=best_k)

# Compute and display the accuracy
num_correct = np.sum(y_test_pred == y_test)
accuracy = float(num_correct) / num_test
print('Got %d / %d correct => accuracy: %f' % (num_correct, num_test, accuracy))
```

k-NN的决策函数包含排序和取最值的操作，是非线性的函数，并且其距离计算函数也不是线性的，即便是1-NN，也是对距离差取了绝对值，因此k-NN的决策边界是非线性的。
#### 2. 1-NN的训练误差将始终低于5-NN。
1-NN的训练误差必然为0，因为离每个训练样本最近的训练数据就是该训练样本本身，因此分类一定是正确的。而对于5-NN，其训练误差极有可能大于0，对于某些数据，可能还不低。但是对于5-NN，一定存在训练数据，使得其训练误差为0，也就是训练数据在5-NN的距离计算模型下，各类别内的距离较之各类别间的距离都非常小，并且每个类别内的数据量都足够，那么此时该训练数据在5-NN模型下的训练误差也可以为0，并不高于1-NN下的训练误差。但是对于一半的训练样本，1-NN的训练误差依旧为0，而5-NN的训练误差极可能非0，对于某些训练数据，5-NN的训练误差还有可能很大。
#### 3. 1-NN的测试误差将始终低于5-NN。
该问题的原因与上一条类似，对于类别内的距离较之各类别间的距离都非常小，并且每个类别内的数据量都足够的训练数据，两者都能做到0的测试误差。同时对于一般的训练数据，1-NN的测试误差也不一定低于5-NN，两个模型计算距离的方式是不同的，有些数据可能更适合L1距离下的分类，有些数据可能更适合L5距离下的分类。
#### 4. 使用k-NN分类器对测试示例进行分类所需的时间随训练集的大小而增加。
对于每个测试示例，要计算其与所有训练样本之间的距离。在距离计算算法相同的情况下，计算所有距离的时间与训练集的大小成正比。因此使用k-NN分类器对测试示例进行分类所需的时间随训练集的增大而增加。
